```python
# 被害者の年齢を年代別にまとめる
# 列名が存在するか確認
age_col = '被害者の年齢'
if age_col not in matsudo_df.columns:
    raise KeyError(f"列が見つかりません: {age_col}")

# 正規表現を使って年代カテゴリに変換
import re
import numpy as np

# 年齢文字列を年代カテゴリに変換する関数
def age_to_group(s):
    if pd.isna(s):
        return '不明/その他'
    s = str(s).strip()
    # 例: '20歳代' の場合そのまま返す
    m = re.match(r"^(\d{1,3})\s*歳代$", s)
    if m:
        decade = int(m.group(1))
        # 例で '20歳代' のように2桁でもOK
        return f"{decade}歳代"
    # '20歳' のような正確な年齢 -> 年代に変換
    m2 = re.search(r"(\d{1,3})\s*歳", s)
    if m2:
        age = int(m2.group(1))
        decade = (age // 10) * 10
        return f"{decade}歳代"
    # 数字だけが入っている場合
    m3 = re.match(r"^\d{1,3}$", s)
    if m3:
        age = int(s)
        decade = (age // 10) * 10
        return f"{decade}歳代"
    # その他表記（例: '不明', 'その他', '10歳未満' など）を分類
    if '不明' in s or s == '' or 'その他' in s:
        return '不明/その他'
    if '未満' in s:
        return '0歳代'  # 未満は0歳代として扱う
    # 最後に保険的に『不明/その他』に振る
    return '不明/その他'

# 新しい列 '年代' を作成
matsudo_df = matsudo_df.copy()
matsudo_df['年代'] = matsudo_df[age_col].apply(age_to_group)

# 年代の出現順（0歳代,0歳代があるか確認, 10歳代,...90歳代, 不明/その他）
ordered = [f"{i}歳代" for i in range(0,100,10)] + ['不明/その他']
# ただし存在しない年代は順序から除外して後でソート可能にする
present = [g for g in ordered if g in matsudo_df['年代'].unique()]

# 年代別の件数をカウントし、存在する年代だけを抽出して順序を保つ
age_counts = matsudo_df['年代'].value_counts().reindex(present, fill_value=0)

# 比率も付ける
total = age_counts.sum()
age_df = (age_counts.rename('件数').to_frame()
          .assign(割合=lambda d: (d['件数'] / total * 100).round(2)))

# 表として表示
print(f"被害者年齢データ件数（松戸市）: {total}")
display(age_df)

# 棒グラフで表示
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(age_df.index.astype(str), age_df['件数'], color='steelblue')
for i, v in enumerate(age_df['件数']):
    ax.text(i, v + max(age_df['件数']) * 0.01, str(int(v)), ha='center', va='bottom')
ax.set_title('松戸市 被害者の年代別 件数', fontsize=14)
ax.set_ylabel('件数')
ax.set_xlabel('年代')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
