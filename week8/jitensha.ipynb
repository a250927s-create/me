```python
# 松戸市だけのデータを取り出す
matsudo_df = df[df['市区町村（発生地）'] == '松戸市']
print(f"松戸市のデータ件数: {len(matsudo_df)}")

matsudo_df.head()
#ranking=matsudo_df.value_counts().sort_index()
```

```python
# 松戸市の町丁目別ランキング（上位10）
# matsudo_df がなければ作る
if 'matsudo_df' not in globals():
    matsudo_df = df[df['市区町村（発生地）'] == '松戸市']

# 町丁目ごとの件数を集計（欠損は '不明' として扱う）
ranking = (
    matsudo_df['町丁目（発生地）'].fillna('不明')
    .value_counts()
    .rename_axis('町丁目')
    .reset_index(name='件数')
    .sort_values('件数', ascending=False)
)

# 棒グラフで可視化（上位10）
import matplotlib.pyplot as plt

top10 = ranking.head(10).set_index('町丁目')
plt.figure(figsize=(10,6))
ax = top10['件数'].plot(kind='bar')
ax.set_title('松戸市 町丁目別 発生件数 上位20')
ax.set_ylabel('件数')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# ランキングのインデックスを1から始める
ranking.index = range(1, len(ranking) + 1)
display(ranking.head(10))
```
